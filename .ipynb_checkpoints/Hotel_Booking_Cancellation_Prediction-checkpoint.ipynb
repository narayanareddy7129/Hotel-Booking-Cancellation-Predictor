### Step 5: Outlier Detection & Treatment

```python
def detect_outliers_iqr(data, column):

    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)

    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = data[
        (data[column] < lower) |
        (data[column] > upper)
    ]

    print(f"{column} Outliers:", outliers.shape[0])

    return lower, upper
```
